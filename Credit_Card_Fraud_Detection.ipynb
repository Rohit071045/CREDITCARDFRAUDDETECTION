```python
# Assuming you are using pandas for X_train and X_test
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Example data (replace X and Y with your actual dataset)
# X_train, X_test, Y_train, Y_test are split from X, Y

# Step 1: Scale the data using StandardScaler
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)

# Step 2: Initialize and train the Logistic Regression model
model = LogisticRegression(max_iter=200, solver='saga')
model.fit(X_train_scaled, Y_train)

# Step 3: Make predictions on the training data
X_train_prediction = model.predict(X_train_scaled)

# Step 4: Evaluate the model on the training set
train_accuracy = accuracy_score(Y_train, X_train_prediction)
print(f"Training Accuracy: {train_accuracy * 100:.2f}%")

# Step 5: Make predictions on the test data
X_test_prediction = model.predict(X_test_scaled)

# Step 6: Evaluate the model on the test set
test_accuracy = accuracy_score(Y_test, X_test_prediction)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```
